```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Cargar los datos
df = pd.read_excel("DoeEIAT.xlsx")

# Dividir los datos en características (X) y variable objetivo (Y)
X = df[["C_HAuCl4*3H2O","C_AgNO3", "V_AgNO3","C*VHCL", "V_HCl","C_NaBH4"]].values
Y = df['LSPR'].values.reshape(-1, 1)
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2021)
```

```python
# Definir el modelo de red neuronal
mlp_model = MLPRegressor(random_state=2021)
```

```python
# Definir el espacio de búsqueda de hiperparámetros #Adam como activación?
param_grid = {
    'hidden_layer_sizes': [(10,7),(9,9),(8,8),(7,7),(6,6),(5,5)],
    'activation': ['relu',"identy"],
    'solver': ['adam', 'RMSprop', "sgd", "lbfgs"],
    "learning_rate":["constant","invscaling","adaptive"],
    'alpha': [0.0001, 0.001, 0.01,0.1, 1],
    "max_iter":[200,500,1000],
    "tol": [1e-4, 1e-3, 1e-2]
}
```

```python
# Crear el objeto GridSearchCV #Probar con menor tamaño (3)
grid_search = GridSearchCV(estimator=mlp_model,
                           param_grid=param_grid,
                           scoring='neg_mean_squared_error', 
                           cv=3)
```

```python
# Ajustar el modelo usando GridSearchCV
grid_search.fit(X_train, y_train)
```

```python
# Obtener los resultados
best_params = grid_search.best_params_
best_score = -grid_search.best_score_
y_pred = grid_search.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
```

```python
# Imprimir los resultados
print("Mejores hiperparámetros:", best_params)
print("Mejor puntuación (neg_mean_squared_error):", best_score)
print("MSE en el conjunto de prueba:", mse)
```
